```python
import pandas as pd
```

```python
employee_data=pd.read_csv("/content/employees.csv")
department_data=pd.read_csv("/content/departments.csv")
```

```python
employee_data
```

```python
department_data
```

*PART A JOINS*

```python
inner_join = pd.merge(employee_data,department_data,on=["Department_ID"],how="inner")
inner_join
```

```python
left_join=pd.merge(employee_data,department_data,on=["Department_ID"],how="left")
left_join
```

```python
right_join=pd.merge(employee_data,department_data,on=["Department_ID"],how="right")
right_join
```

```python
full_outer_join=pd.merge(employee_data,department_data,on=["Department_ID"],how="outer")
full_outer_join
```

*PART B Group By*

```python
Sum=employee_data.groupby("Department_ID")["Salary"].agg(['sum']).reset_index()
Sum
```

```python
Average=employee_data.groupby("Department_ID")["Salary"].agg(["mean"]).reset_index()
Average
```

```python
grp_by_loc= pd.merge(employee_data, department_data, on='Department_ID', how='inner')
data = grp_by_loc.groupby('Location').agg(
    No_of_emp=('Employee_ID', 'count'),
    Total_salary=('Salary', 'sum')).reset_index()
data
```

```python
Filter_emp= employee_data[employee_data['Salary'] > 50000]
avg_sal = Filter_emp.groupby('Department_ID')['Salary'].mean().reset_index()
avg_sal
```
