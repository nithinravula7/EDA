```python
import pandas as  pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("//media/sales_dataset.csv")
```

```python
data
```

```python
# Identify columns with missing values and their counts
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Impute missing values in 'Region'
data['Region'].fillna('Unknown', inplace=True)

# Impute missing values in 'Units_Sold' using the median
data['Units_Sold'].fillna(data['Units_Sold'].median(), inplace=True)
```

```python
data
```

```python
# Calculating quartiles and IQR
Q1 = data['Units_Sold'].quantile(0.25)
Q3 = data['Units_Sold'].quantile(0.75)
IQR = Q3 - Q1

# Calculating lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# Identifying outliers
outliers = data[(data['Units_Sold'] < lower_bound) | (data['Units_Sold'] > upper_bound)]
```

```python
# Removing outliers
cleaned_data = data[(data['Units_Sold'] >= lower_bound) & (data['Units_Sold'] <= upper_bound)]

# Comparring sizes
print("Original Dataset Size:", data.shape)
print("Cleaned Dataset Size:", cleaned_data.shape)
```

```python


# Box plot before outlier treatment
plt.figure(figsize=(8, 6))
plt.boxplot(data['Units_Sold'])
plt.title('Units_Sold - Before Outlier Treatment')
plt.ylabel('Units Sold')
plt.show()

# Box plot after outlier treatment
plt.figure(figsize=(8, 6))
plt.boxplot(cleaned_data['Units_Sold'])
plt.title('Units_Sold - After Outlier Treatment')
plt.ylabel('Units Sold')
plt.show()


```
