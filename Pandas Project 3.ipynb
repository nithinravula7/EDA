```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# ***Loading & Exploring data***

```python
d1=pd.read_csv("/content/amsterdam_weekdays.csv")
d2=pd.read_csv("/content/amsterdam_weekends.csv")
d3=pd.read_csv("/content/athens_weekdays.csv")
d4=pd.read_csv("/content/athens_weekends.csv")
d5=pd.read_csv("/content/barcelona_weekdays.csv")
d6=pd.read_csv("/content/barcelona_weekends.csv")
d7=pd.read_csv("/content/berlin_weekdays.csv")
d8=pd.read_csv("/content/berlin_weekends.csv")
d9=pd.read_csv("/content/budapest_weekdays.csv")
d10=pd.read_csv("/content/budapest_weekends.csv")
d11=pd.read_csv("/content/lisbon_weekdays.csv")
d12=pd.read_csv("/content/lisbon_weekends.csv")
d13=pd.read_csv("/content/london_weekdays.csv")
d14=pd.read_csv("/content/london_weekends.csv")
d15=pd.read_csv("/content/paris_weekdays.csv")
d16=pd.read_csv("/content/paris_weekends.csv")
d17=pd.read_csv("/content/rome_weekdays.csv")
d18=pd.read_csv("/content/rome_weekends.csv")
d19=pd.read_csv("/content/vienna_weekdays.csv")
d20=pd.read_csv("/content/vienna_weekends.csv")
```

```python

d1["city"]="Amsterdam"
d3["city"]="Athens"
d5["city"]="Berlin"
d7["city"]="Budapest"
d9["city"]="Barcelona"
d11["city"]="Lisbon"
d13["city"]="London"
d15["city"]="Paris"
d17["city"]="Rome"
d19["city"]="Vienna"
d2["city"]="Amsterdam"
d4["city"]="Athens"
d6["city"]="Berlin"
d8["city"]="Budapest"
d10["city"]="Barcelona"
d12["city"]="Lisbon"
d14["city"]="London"
d16["city"]="Paris"
d18["city"]="Rome"
d20["city"]="Vienna"
```

```python
weekdays=pd.concat([d1,d3,d5,d7,d9,d11,d13,d15,d17,d19],ignore_index=True)
weekdays.head(5)
```

```python

weekends=pd.concat([d2,d4,d6,d8,d10,d12,d14,d16,d18,d20],ignore_index=True)
weekends.head(5)
```

```python
weekends["data_type"]="weekend"
weekdays["data_type"]="weekdays"
```

```python
combined_data=pd.concat([weekdays,weekends],ignore_index=True)
combined_data.head(5)
```

# ***Data Cleaning***

```python
print(combined_data.isnull().sum())
```

```python
print(combined_data.duplicated().sum())
```

```python
print(combined_data.dtypes)
```

```python
important_columns = ['realSum', 'person_capacity', 'guest_satisfaction_overall', 'bedrooms']
combined_data[important_columns].plot(kind='box', subplots=True, layout=(2, 2), figsize=(12, 8))

plt.show()
```

# ***Univariate Analysis***

```python
plt.figure(figsize=(10, 5))
plt.hist(combined_data[combined_data['data_type'] == 'weekdays']['realSum'], bins=20, alpha=0.5, label='Weekdays')
plt.hist(combined_data[combined_data['data_type'] == 'weekend']['realSum'], bins=20, alpha=0.5, label='Weekends')
plt.title('Price Distribution (Weekdays vs. Weekends)')
plt.xlabel('Price (realSum)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='data_type', y='realSum', data=combined_data)
plt.title('Price Distribution (Weekdays vs. Weekends)')
plt.xlabel('Data Type')
plt.ylabel('Price (realSum)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='room_type', data=combined_data)
plt.title('Room Type Counts')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(combined_data['guest_satisfaction_overall'], bins=20)
plt.title('Guest Satisfaction Distribution')
plt.xlabel('Guest Satisfaction Rating')
plt.ylabel('Frequency')
plt.show()
```

# ***Bivariate Analysis***

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='room_type', y='realSum', hue='data_type', data=combined_data)
plt.title('Average Price by Room Type (Weekdays vs. Weekends)')
plt.xlabel('Room Type')
plt.ylabel('Average Price (realSum)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='host_is_superhost', y='realSum', data=combined_data)
plt.title('Price Distribution by Superhost Status')
plt.xlabel('Superhost Status')
plt.ylabel('Price (realSum)')
plt.show()
plt.figure(figsize=(10, 6))
sns.boxplot(x='host_is_superhost', y='guest_satisfaction_overall', data=combined_data)
plt.title('Guest Satisfaction by Superhost Status')
plt.xlabel('Superhost Status')
plt.ylabel('Guest Satisfaction Rating')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='dist', y='realSum', data=combined_data)
plt.title('Price vs. Distance from City Center')
plt.xlabel('Distance from City Center (dist)')
plt.ylabel('Price (realSum)')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='metro_dist', y='realSum', data=combined_data)
plt.title('Price vs. Distance from Metro Station')
plt.xlabel('Distance from Metro Station (metro_dist)')
plt.ylabel('Price (realSum)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='rest_index', y='guest_satisfaction_overall', data=combined_data)
plt.title('Guest Satisfaction vs. Restaurant Index')
plt.xlabel('Restaurant Index')
plt.ylabel('Guest Satisfaction')
plt.show()
```

```python
correlation = combined_data[['realSum', 'guest_satisfaction_overall', 'attr_index', 'rest_index']].corr()
print(correlation)
```

```python
city_prices = combined_data.groupby(['city', 'data_type'])['realSum'].agg(['mean', 'median'])
print(city_prices)
```

```python
city_prices = city_prices.reset_index()  # Reshape data for plotting

plt.figure(figsize=(12, 6))
sns.barplot(x='city', y='mean', hue='data_type', data=city_prices)
plt.title('Average Price by City (Weekdays vs. Weekends)')
plt.xlabel('City')
plt.ylabel('Average Price')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Data Type')
plt.show()
```

```python
room_capacity = combined_data.groupby(['city', 'data_type'])['person_capacity'].mean()
print(room_capacity)
```

```python
room_capacity = room_capacity.reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='city', y='person_capacity', hue='data_type', data=room_capacity)
plt.title('Average Room Capacity by City (Weekdays vs. Weekends)')
plt.xlabel('City')
plt.ylabel('Average Room Capacity')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Data Type')
plt.show()
```

```python
satisfaction_scores = combined_data.groupby(['city', 'data_type'])['guest_satisfaction_overall'].agg(['mean', 'std'])
print(satisfaction_scores)
```

```python
satisfaction_scores = satisfaction_scores.reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='city', y='mean', hue='data_type', data=satisfaction_scores)
plt.title('Average Guest Satisfaction by City (Weekdays vs. Weekends)')
plt.xlabel('City')
plt.ylabel('Average Guest Satisfaction')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Data Type')
plt.show()
```

# ***Insights and Recommendaions***


**For Hosts:**

*Pricing:* Adjust prices based on city, day of the week (weekends are usually higher), room type, and local attractiveness.

*Guest Satisfaction:* Keep your place clean, provide amenities, respond quickly to guests, and become a Superhost.

**For Guests:**

*Lower Prices:* Travel during weekdays, consider less popular cities, and be flexible with your dates and room type.

*Higher Satisfaction:* Choose listings from Superhosts, read reviews, and pick a place with the amenities and location you want.
