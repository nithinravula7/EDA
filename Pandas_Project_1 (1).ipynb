***Handling Missing Vaues***

```python
cars_data['series'] = pd.to_numeric(cars_data['series'], errors='coerce')
cars_data['series'] = cars_data['series'].fillna(cars_data['series'].mean())
cars_data['transmission'] = pd.to_numeric(cars_data['transmission'], errors='coerce')
cars_data['transmission'] = cars_data['series'].fillna(cars_data['transmission'].mean())


```

## ***14. Business Insights***
Based on the analysis performed in the previous sections, here's a summary of the factors that seem to have the most significant influence on the price of a used car:

**Make and Model**: Certain car makes and models tend to hold their value better than others, resulting in higher prices even when other factors are similar. This was evident in the analysis of average prices for different makes and models (Section 7).

**Kilometers Driven**: As expected, there's a negative correlation between the number of kilometers driven and the price of a used car. Higher mileage generally leads to lower prices (Section 6).

**Car Age**: The age of the car is another important factor. Newer cars tend to command higher prices, while older cars depreciate in value (Section 13).

**Fuel Type**: The type of fuel used by the car can also influence its price. Diesel cars often have a higher resale value compared to petrol cars (Section 7).

**Location**: Prices can vary depending on the city or state where the car is being sold. Demand and supply dynamics in different locations can affect prices (Section 11).
